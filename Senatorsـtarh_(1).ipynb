add nodes and edges

```python
from networkx.algorithms import bipartite
bills = set()
for bill in df['tarh_emza_shedeh']:
    bill_list =str(bill).split(',')
    bills.update(bill_list)

senators=tuple(df['name'])
G=nx.Graph()
G.add_nodes_from(senators, bipartite=0)
G.add_nodes_from(bills, bipartite=1)
#G.nodes
```

```python
for _, row in df.iterrows():
    representatives = row['name']
    bills = str(row['tarh_emza_shedeh']).split(',')

    for bill in bills:
        G.add_edge(representatives, bill)
```

visualization

```python
pos2=nx.bipartite_layout(G , nodes=df['name'])
#pos2=nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G , pos2 , nodelist=df['name'] , node_size= 50, node_color='blue')
nx.draw_networkx_nodes(G , pos2 , nodelist=bills , node_size= 50, node_color='red')

nx.draw_networkx_edges(G , pos2, arrowsize=2)
#nx.draw_networkx_labels(G , pos2, font_size=6)


plt.axis('off')
plt.show()
```

```python
file_path='/Users/fatemeh/Desktop/graph'
nx.write_graphml(G,'senators_law.graphml')
```

```python
projected_graph = nx.bipartite.projected_graph(G, nodes=df['name'])
nx.write_graphml(G,'senators_law_onepartite.graphml')

```

**greedy_modularity**

```python
from networkx.algorithms import community

communities = list(community.greedy_modularity_communities(projected_graph))
num_communities = len(communities)
print("Number of Communities:", num_communities)

for i, community_nodes in enumerate(communities):
    print(f"Community {i + 1}: {list(community_nodes)}")
```

```python
from networkx.algorithms.community.louvain import modularity
greedy_modularity=modularity(projected_graph ,communities )
greedy_modularity
```

louvain

```python
from networkx.algorithms import community

louvain_communities = list(community.louvain_communities(projected_graph , resolution=1))
num_communities = len(louvain_communities)
print("Number of Communities:", num_communities)

for i, community_nodes in enumerate(louvain_communities):
    print(f"Community {i + 1}: {list(community_nodes)}")
```

```python
from networkx.algorithms.community.louvain import modularity
greedy_modularity=modularity(projected_graph ,louvain_communities )
greedy_modularity
```

girvan_newman

```python
from networkx.algorithms import community

gn_communities = list(community.girvan_newman(projected_graph))
num_communities = len(gn_communities)
print("Number of Communities:", num_communities)

for i, community_nodes in enumerate(gn_communities):
    print(f"Community {i + 1}: {list(community_nodes)}")
```
